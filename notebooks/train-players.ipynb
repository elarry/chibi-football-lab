# Training Deep RL Models in Multi-Agent settings with ML-Agents


## ML-Agents Basics

Basic command follows form:
```bash
mlagents-learn $CONFIG_PATH \
    --env=$UNITY_ENV \
    --run-id=$RUN_ID \
    --no-graphics
```

Useful additional arguments (see `mlagents-learn --help` for more), include:
- `--num-envs 8`: Use 8 concurrent Unity environment instances to use for training.
- `--baseport 5006`: If training is already runnning, need to assign different port for new independent concurrent run (e.g. port 5006).
- `--no-graphics`: Remove this to see training behaviour.
- `--resume`: Resume training from existing preliminary results from `$RUN_ID`. Note that the ELO counter is reset to 1200 upon resumption.
- `--force`: Overwrite any results with same `$RUN_ID`
- `--initialize-from $RUN_ID_PREVIOUS`: Start training from `$RUN_ID_PREVIOUS`, save results specified in `--run-id $RUN_ID_NEW`
